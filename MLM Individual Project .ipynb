-   **K-Nearest Neighbors (KNN)**: A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in nm26_non_cat_vars:
    ci_results[col] = confidence_interval(nm26_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

nm26_contingency_table = pd.crosstab(nm26_sample['Import_Export'], nm26_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(nm26_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
nm26_alpha = 0.05  # Significance level

if p < nm26_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    nm26_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    nm26_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    nm26_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    nm26_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    nm26_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return nm26_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in nm26_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(nm26_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(nm26_non_cat_vars) > 1:
        for i in range(len(nm26_non_cat_vars)):
            for j in range(i+1, len(nm26_non_cat_vars)):
                col1, col2 = nm26_non_cat_vars[i], nm26_non_cat_vars[j]
                t_test_result = t_test_correlation(nm26_sample[col1], nm26_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```


**Observations for Normality Tests for Quantity:**

# Model Run Statistics

## Time Taken

```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55026)
log_reg_time = time_model_run(log_reg_model, nm26_X_train, nm26_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55026)
svm_time = time_model_run(svm_model, nm26_X_train, nm26_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55026)
sgd_time = time_model_run(sgd_model, nm26_X_train, nm26_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55026)
dt_time = time_model_run(dt_model, nm26_X_train, nm26_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, nm26_X_train, nm26_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, nm26_X_train, nm26_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55026)
rf_time = time_model_run(rf_model, nm26_X_train, nm26_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55026)
xgb_time = time_model_run(xgb_model, nm26_X_train, nm26_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```


**Observations:**

-   **Fastest:** KNN and Naive Bayes are the fastest, taking 0.0054 seconds and 0.0051 seconds, respectively.
-   **Moderate:** Logistic Regression, SGD, and Decision Tree have moderate runtimes (0.0179 seconds, 0.0318 seconds, and 0.0316 seconds, respectively).
-   **Relatively Slow:** XGBoost has a relatively slow runtime, taking 0.2889 seconds.
-   **Slowest:** SVM and Random Forest are the slowest, taking 0.5545 seconds and 1.2530 seconds, respectively.

**Insights:**

-   **KNN and Naive Bayes** are the fastest, ideal for quick analysis or when computational resources are limited.
-   **Logistic Regression, SGD, and Decision Tree** offer a good balance of speed and complexity, making them suitable for many applications where quick results are desired.
-   **XGBoost** has a moderate runtime, offering a balance between speed and accuracy. It might be suitable for tasks where higher accuracy is important but not at the expense of significant computational resources.
-   **SVM and Random Forest** have slower runtimes, requiring more resources. They might be more suitable for tasks where accuracy is paramount and computational resources are not a major constraint, such as high-stakes applications where the cost of misclassification is high.


## Memory Utilized

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55026)
log_reg_memory = get_model_memory_usage(log_reg_model, nm26_X_train, nm26_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55026)
svm_memory = get_model_memory_usage(svm_model, nm26_X_train, nm26_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55026)
sgd_memory = get_model_memory_usage(sgd_model, nm26_X_train, nm26_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55026)
dt_memory = get_model_memory_usage(dt_model, nm26_X_train, nm26_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, nm26_X_train, nm26_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, nm26_X_train, nm26_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55026)
rf_memory = get_model_memory_usage(rf_model, nm26_X_train, nm26_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55026)
xgb_memory = get_model_memory_usage(xgb_model, nm26_X_train, nm26_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```
